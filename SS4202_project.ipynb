Visualisations

Star Type (Red Dwarf, Brown Dwarf, White Dwarf, Main Sequence , SuperGiants, HyperGiants)

```python
fig = px.bar(dataset, y = 'Spectral Class',color="Star type", template ="seaborn")
fig.show()
# fig.write_image("my_figure.png")
```

```python
fig = px.histogram(dataset, y="Spectral Class", x="Star type", color="Temperature (K)",
                   hover_data=dataset.columns, template = 'seaborn', title = 'Spectral Class and Star Type with their temperatures')
fig.show()
# fig.write_image("my_figure.png")
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(y="Spectral Class", data=dataset, palette='flare',);
# plt.savefig('spectral_classes.png', dpi=100, bbox_inches='tight')
```

```python
fig = px.scatter(dataset,
                 x="Star type", y='Absolute magnitude(Mv)', size="Radius(R/Ro)", color="Temperature (K)",
                 log_x=True, size_max=60,
                 template='plotly_dark', title='Star Type vs Absolute Magnitude')
fig.show()
# fig.write_image("my_figure.png")
```

```python
fig = px.density_heatmap(dataset, x="Radius(R/Ro)", y ="Luminosity(L/Lo)",title = 'Radius vs Luminosity', template = 'plotly_dark')
fig.show()
# fig.write_image("my_figure.png")
```

From above plot we can conclude that Luminosity is directly proportional to Radii of Star

```python
fig = px.density_heatmap(dataset, x="Temperature (K)", y ="Absolute magnitude(Mv)",title = 'Temperature vs Absolute Magnitude', template = 'plotly_dark')
fig.show()
# fig.write_image("my_figure.png")
```

From above plot we can conclude that Temperature is not Linearly proportional to Their Magnitude.

```python
fig = px.scatter_ternary(dataset, a="Absolute magnitude(Mv)", b="Luminosity(L/Lo)", c="Radius(R/Ro)",hover_name="Star type",
    color="Temperature (K)", template = 'plotly_dark',size="Temperature (K)", size_max=30,)
fig.show()
# fig.write_image("my_figure.png")
```

From above plot we can conclude that stars having negative Absolute Magnitude have biggest Radii and Highest Luminosity and Temperature.

```python
fig = px.scatter_3d(dataset, x='Star type', y='Absolute magnitude(Mv)', z='Spectral Class',
              color='Temperature (K)', template = 'plotly_dark', title = 'Distribution of Star type and Spectral Class with their Magnitudes')
fig.show()
# fig.write_image("my_figure.png")
```

From Above plot we can coclude that Stars having Spectral Class O tends to have highest Temperature.

```python
fig = px.box(dataset, x="Star type", y="Temperature (K)", template = 'plotly_dark', 
             title = 'Representation of Type of Stars with Temperature')
fig.show()
# fig.write_image("my_figure.png")
```

From Above plot we can conclude
*   Red Dwarfs (Type 0) have lowest Temperature.
*   SuperGiants (Type 5) have highest Temperature.

```python
fig = px.pie(dataset, names='Spectral Class', values='Temperature (K)', template = 'plotly_dark', 
             title = 'Representation of Spectral Classes with their Temperature')
fig.show()
# fig.write_image("my_figure.png")
```

```python
fig = px.imshow(dataset.corr(), template = 'plotly_dark')
fig.show()
# fig.write_image("my_figure.png")
```

From above Correlation Heatmap we can conclude:


*  Temperature and Luminosity have High Positive Correlation.
*   Temperature and Absolute Magnitude have High Negative Correlation.
